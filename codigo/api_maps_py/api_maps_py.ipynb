```python
def formatar_lat_long(valor):
    str_valor = str(valor)

    str_valor = str_valor.replace('.', '')

    if len(str_valor) >= 3:
      str_valor = str_valor[:3] + '.' + str_valor[3:]

    return str_valor

servicosCu['latitude'] = servicosCu['latitude'].apply(formatar_lat_long)
servicosCu['longitude'] = servicosCu['longitude'].apply(formatar_lat_long)
```

```python
import pandas as pd
import googlemaps

# Supondo que 'gmaps' já esteja definido como uma instância válida do Google Maps API.

def obter_distancias(origens, destinos):
    matriz_distancias = gmaps.distance_matrix(origens, destinos, mode='driving', units='metric')
    return matriz_distancias

def dividir_em_lotes(lista, tamanho_lote):
    for i in range(0, len(lista), tamanho_lote):
        yield lista[i:i + tamanho_lote]

# Substitua 'df_origem' e 'df_destino' pelos seus DataFrames reais
df_origem = pd.DataFrame(filtro_funcionarios)  # Substitua os pontos suspensos pelos seus dados reais
df_origem = df_origem.head(20)
df_destino = pd.DataFrame(filtro_demanda)  # Substitua os pontos suspensos pelos seus dados reais
df_destino = df_destino.head(20)

# Inicializar a lista de distâncias
distancias = []

# Dividir em lotes e obter distâncias para cada lote
for lote_origens in dividir_em_lotes(df_origem.to_dict('records'), 10):
    coordenadas_origens = ["{},{}".format(location['Latitude'], location['Longitude']) for location in lote_origens]

    for lote_destinos in dividir_em_lotes(df_destino.to_dict('records'), 10):
        coordenadas_destinos = ["{},{}".format(location['Latitude'], location['Longitude']) for location in lote_destinos]

        # Obter distâncias para o lote atual
        matriz_distancias = obter_distancias(coordenadas_origens, coordenadas_destinos)

        # Construir tabela de distâncias para o lote atual
        for i, origem in enumerate(lote_origens):
            for j, destino in enumerate(lote_destinos):
                try:
                    distancia_km = matriz_distancias['rows'][i]['elements'][j]['distance']['text']
                    tempo_de_deslocamento = matriz_distancias['rows'][i]['elements'][j]['duration']['text']

                    # Adicionar 'Tipo' e 'Setor' à lista de distâncias
                    distancias.append({
                        'Tipo Origem': origem['Tipo'],
                        'Setor Origem': origem['Setor'],
                        'Coordenadas Origem': coordenadas_origens[i],
                        'Tipo Destino': destino['Tipo'],
                        'Coordenadas Destino': coordenadas_destinos[j],
                        'Setor Destino': destino['Setor'],
                        'Distancia': distancia_km,
                        'Tempo de Deslocamento': tempo_de_deslocamento,
                    })
                except KeyError:
                    print(f"Erro ao processar distância entre {origem} e {destino}")

# Criar DataFrame com todas as distâncias
df_distancias = pd.DataFrame(distancias)

# Exibir as primeiras linhas do DataFrame
df_distancias
```
